What are the differences between np.array, np.asarray, and np.asanyarray?

Answer:  
np.array(): This function always creates a new array, copying the input data by default. It’s used when you want a strict NumPy array, even if the input is already an array-like object.

np.asarray(): This function converts the input to a NumPy array. However, if the input is already a NumPy array, it does not copy it, avoiding unnecessary memory usage.

np.asanyarray(): Similar to np.asarray(), but it passes the input unchanged if it’s already a subclass of ndarray (like a matrix or masked array). It's used when you don’t want to force a strict NumPy array but still want to handle array-like objects.

3 . What is the difference between a deep copy and a shallow copy in Python?

Answer:

Shallow Copy: It creates a new object but does not create copies of objects inside it. It only references the objects in the original container, meaning changes to the internal objects affect both the original and the copied object.

Deep Copy: It creates a new object and recursively copies all objects inside the original, meaning the original and copied object are completely independent, and changes to one do not affect the other.

5.How can you create a NumPy array with random integers between 1 and 10 of shape (5, 6) and extract even and odd integers?

```python
# Create a random integer array
random_int_array = np.random.randint(1, 11, size=(5, 6))

# Extract even integers
even_integers = random_int_array[random_int_array % 2 == 0]

# Extract odd integers
odd_integers = random_int_array[random_int_array % 2 != 0]

print(f"Even Integers: {even_integers}")
print(f"Odd Integers: {odd_integers}")
```

6.How can you create a 3D NumPy array of random integers between 1 and 10 and perform operations like finding indices of max values and element-wise multiplication?

```python
# Create a 3D array of random integers between 1 and 10
array_3D = np.random.randint(1, 11, size=(3, 3, 3))

# Find the indices of the maximum values along the third axis
max_indices = np.argmax(array_3D, axis=2)

# Perform element-wise multiplication of the array by 2
multiplied_array = array_3D * 2

print(f"3D Array:\n{array_3D}")
print(f"Indices of max values along depth (third axis):\n{max_indices}")
print(f"Element-wise multiplied array:\n{multiplied_array}")
```

Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib: right y-axis for 'Humidity'). b) Label the x-axis as 'Date'. a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and c) Set the title of the plot as 'Temperature and Humidity Over Time

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a time-series dataset with columns 'Date', 'Temperature', and 'Humidity'
date_range = pd.date_range(start='2023-01-01', periods=30, freq='D')
temperature = np.random.uniform(15, 35, size=30)  # Random temperatures between 15 and 35
humidity = np.random.uniform(40, 80, size=30)  # Random humidity between 40 and 80

# Create a DataFrame
df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Step 2: Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Step 3: Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib: a) Plot a histogram of the data with 30 bins. b) Overlay a line plot representing the normal distribution's probability density function (PDF). c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'. d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate 1000 samples from a normal distribution
data = np.random.randn(1000)

# Step 2: Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Data Histogram')

# Step 3: Overlay a line plot representing the normal distribution's PDF
xmin, xmax = plt.xlim()  # Get the current x-axis limits
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))  # Calculate the PDF
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Step 4: Label the axes and set the title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Step 5: Display the plot
plt.show()
```

Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generating two random arrays for the scatter plot
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Assigning quadrants based on the points' position relative to the origin
quadrants = []
for i in range(len(x)):
    if x[i] >= 0 and y[i] >= 0:
        quadrants.append('Quadrant I')
    elif x[i] < 0 and y[i] >= 0:
        quadrants.append('Quadrant II')
    elif x[i] < 0 and y[i] < 0:
        quadrants.append('Quadrant III')
    else:
        quadrants.append('Quadrant IV')

# Creating the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, hue=quadrants, palette="Set1")

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Displaying the plot with a legend
plt.legend(title='Quadrants')
plt.show()
```

18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

```python
from bokeh.plotting import figure, output_notebook, show
import numpy as np

# Setting up Bokeh to display in the notebook
output_notebook()

# Generate x values and compute the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a new plot, replacing 'plot_width' and 'plot_height' with 'width' and 'height'
p = figure(title="Sine Wave Function",
           x_axis_label='X-axis',
           y_axis_label='Y-axis',
           width=700, height=400) # Changed plot_width/height to width/height

# Add a line renderer with legend and line thickness
p.line(x, y, line_width=2)

# Enable grid lines
p.xgrid.visible = True
p.ygrid.visible = True

# Display the plot
show(p)
```

19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical bar chart

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

# Set up Bokeh to display in notebook
output_notebook()

# Randomly generate categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)

# Create a DataFrame to store the categorical data
df = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(df)

# Create a color mapper based on the 'values'
mapper = linear_cmap(field_name='values', palette=Viridis256, low=min(values), high=max(values))

# Create the bar chart
# Changed 'plot_height' and 'plot_width' to 'height' and 'width' respectively
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values',
           height=400, width=700, tools="")

# Add bars to the plot
p.vbar(x='categories', top='values', width=0.8, source=source,
       line_color='white', fill_color=mapper)

# Add hover tool to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Display the plot
show(p)
```

20.Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)
y = np.random.randn(100)

# Create the line plot
fig = go.Figure()

# Add a trace for the line
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Line'))

# Set the title and labels
fig.update_layout(title='Simple Line Plot',
                  xaxis_title='X-axis',
                  yaxis_title='Y-axis')

# Show the plot
fig.show()
```

21.Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)

# Create the pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hoverinfo='label+percent', textinfo='value+percent')])

# Set the title
fig.update_layout(title_text='Interactive Pie Chart')

# Show the plot
fig.show()
```
